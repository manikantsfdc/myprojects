```python
rapedata=rapedata[rapedata['Subgroup'] !='Total Rape Victims']
```

```python
# check if the all the rape cases are reported
rapedata[rapedata['Victims_of_Rape_Total'] != rapedata['Rape_Cases_Reported']].head()
```

```python
#as there are unreported case too , we will create a new column of unreported cases 
rapedata['Unreported_Cases'] = rapedata['Victims_of_Rape_Total'] - rapedata['Rape_Cases_Reported']
```

```python
rapedata[rapedata['Unreported_Cases'] > 0].head()
```

```python
# We will now plot the unreported rape cases sorted by states throughout 2001 to 2010
unreported_victims_by_state = rapedata.groupby('Area_Name').sum()
unreported_victims_by_state.drop('Year', axis = 1, inplace = True)
```

```python
#Plotting 
plt.subplots(figsize = (15, 8))
ab = unreported_victims_by_state[unreported_victims_by_state['Unreported_Cases'] 
                                 > 0]['Unreported_Cases'].sort_values(ascending = False)

#print(ct)
cd = ab.plot.bar()
cd.set_xlabel('Indian States')
cd.set_ylabel('Total No. of Unreported Rape Victims from 2001 to 2010')
cd.set_title('Statewise total Unreported Rape Victims from 2001 to 2010')
plt.show()
```

```python
rape_victims_by_state = rapedata.groupby('Area_Name').sum()
rape_victims_by_state.drop('Year', axis = 1, inplace = True)
print('Total Rape Victims = ' ,rape_victims_by_state['Rape_Cases_Reported'].sum())
rape_victims_by_state.sort_values(by = 'Rape_Cases_Reported', ascending = False).head()
```

```python
#lets plot only the total number of rape cases reported in each state
plt.subplots(figsize = (15, 6))
ef = rape_victims_by_state['Rape_Cases_Reported'].sort_values(ascending = False)
#print(ef)
gh = ef.plot.bar()
#ax = sns.barplot(x = rape_victims_by_state.index, y = rape_victims_by_state['Rape_Cases_Reported'])
gh.set_xlabel('States')
gh.set_ylabel('Total No. of Reported Rape Victims from 2001 to 2010')
gh.set_title('Statewise total Reported Rape Victims from Year 2001 to 2010')
plt.show()
print(ef)
```

```python
#as mp has highest number, lets look into that further,
rape_victims_mp = rapedata[rapedata['Area_Name'] == 'Madhya Pradesh']

# top details 
rape_victims_mp.head()
```

```python
rape_victims_by_year_mp = rape_victims_mp.groupby('Year').sum()
# plotting the data
plt.subplots(figsize = (15, 6))
ij = rape_victims_by_year_mp['Rape_Cases_Reported'].plot()
ij.xaxis.set_ticks(np.arange(2001, 2011, 1))
ij.set(xlabel = 'Year', ylabel = 'Total No. of Reported Rape Victims', 
       title = 'No. of Rape Victims frome year 2001 to 2010 of Madhya Pradesh')
plt.show()
```
