```python
dt= data2019[(data2019['Country or region'].isin(['India','Canada','United Kingdom', 'United States']))]
dt
```

```python
ax = dt.plot(y="Social support", x="Country or region", kind="bar",color='C3')
dt.plot(y="GDP per capita", x="Country or region", kind="bar", ax=ax, color="C1")
dt.plot(y="Healthy life expectancy", x="Country or region", kind="bar", ax=ax, color="C2")

plt.show()
```

```python
ax = dt.plot(y="Freedom to make life choices", x="Country or region", kind="bar",color='C3')
dt.plot(y="Generosity", x="Country or region", kind="bar", ax=ax, color="C1",)
dt.plot(y="Perceptions of corruption", x="Country or region", kind="bar", ax=ax, color="C2",)

plt.show()
```

```python
import plotly.graph_objs as go
from plotly.offline import iplot

data = dict(type = 'choropleth', 
           locations = data2019['Country or region'],
           locationmode = 'country names',
           colorscale='RdYlGn',
           z = data2019['Score'], 
           text = data2019['Country or region'],
           colorbar = {'title':'Happiness Score'})

layout = dict(title = 'Geographical Visualization of Happiness Score for year 2019', 
              geo = dict(showframe = True, projection = {'type': 'azimuthal equal area'}))

choromap3 = go.Figure(data = [data], layout=layout)
iplot(choromap3)
```

```python
plt.figure(figsize=(10,5))
sb.kdeplot(data2015['Health (Life Expectancy)'],color='red')
sb.kdeplot(data2016['Health (Life Expectancy)'],color='blue')
sb.kdeplot(data2017['Health..Life.Expectancy.'],color='limegreen')
sb.kdeplot(data2018['Healthy life expectancy'],color='orange')
sb.kdeplot(data2019['Healthy life expectancy'],color='pink')
plt.title('Health over the Years',size=20)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sb.kdeplot(data2015['Economy (GDP per Capita)'],color='red')
sb.kdeplot(data2016['Economy (GDP per Capita)'],color='blue')
sb.kdeplot(data2017['Economy..GDP.per.Capita.'],color='limegreen')
sb.kdeplot(data2018['GDP per capita'],color='orange')
sb.kdeplot(data2019['GDP per capita'],color='pink')
plt.title('Economy over the Years',size=20)
plt.show()
```
